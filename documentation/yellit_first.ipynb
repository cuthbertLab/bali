```python
import io
with io.open('all_patterns.txt', encoding='utf-8') as ap:
    all_patterns = ap.readlines()

for i in range(len(all_patterns)):
    all_patterns[i] = all_patterns[i].strip()
```

```python
all_patterns[0:12]
```

```python
strokePatterns = []
for p in all_patterns:
    if p.startswith(u'(r)'):
        strokePatterns.append(p)
```

```python
strokePatterns
```

```python
numOfMatches = 0
for thisStrokePattern in strokePatterns:
    if 'T r e e T r e e T r e e' in thisStrokePattern:
        numOfMatches = numOfMatches + 1 
        print(thisStrokePattern)
print(numOfMatches)
```
